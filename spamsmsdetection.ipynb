```python
import opendatasets as od
od.download("https://www.kaggle.com/datasets/uciml/sms-spam-collection-dataset")
```

```python
import pandas as pd
data=pd.read_csv("/content/sms-spam-collection-dataset/spam.csv",encoding="ISO-8859-1")
data.rename(columns={"v1":"Status",
                     "v2":"Text1",
                     "Unnamed: 2":"Text2",
                     "Unnamed: 3":"Text3",
                     "Unnamed: 4":"Text4"},inplace=True)
data
```

```python
data.isnull().sum()
```

```python
data['Text']=data['Text1'].fillna("").astype(str)+" "+data['Text2'].fillna("").astype(str)+" "+data['Text3'].fillna("").astype(str)+" "+data['Text4'].fillna("").astype(str)+" "
data=data.drop(columns=['Text1','Text2','Text3','Text4'])
data
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
data.describe()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data['Status']=le.fit_transform(data['Status'])
data
```

```python
x=data['Text']
y=data['Status']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(stop_words='english',max_features=5000)
x_train_tfidf=vectorizer.fit_transform(x_train)
x_test_tfidf=vectorizer.transform(x_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
nb_model=MultinomialNB()
nb_model.fit(x_train_tfidf,y_train)
y_pred_nb=nb_model.predict(x_test_tfidf)
```

```python
from sklearn.metrics import classification_report,accuracy_score
print("Naive Bayes Classifier:")
print("Accuracy:",accuracy_score(y_test,y_pred_nb))
print(classification_report(y_test,y_pred_nb))
```

```python
from sklearn.linear_model import LogisticRegression
lr_model=LogisticRegression()
lr_model.fit(x_train_tfidf,y_train)
y_pred_lr=lr_model.predict(x_test_tfidf)
```

```python
from sklearn.metrics import classification_report,accuracy_score
print("\nLogistic Regression Classifier:")
print("Accuracy:",accuracy_score(y_test,y_pred_lr))
print(classification_report(y_test,y_pred_lr))
```

```python
from sklearn.svm import SVC
svm_model=SVC()
svm_model.fit(x_train_tfidf,y_train)
y_pred_svm=svm_model.predict(x_test_tfidf)
```

```python
from sklearn.metrics import classification_report,accuracy_score
print("\nSupport Vector Machine Classifier:")
print(classification_report(y_test,y_pred_svm))
print("Accuracy:",accuracy_score(y_test,y_pred_svm))
```
